```python
from autoscraper import AutoScraper
```

```python
yoshop_url="https://yoshops.com/t/headphones"

wanted_list=["₹799.00","Boat Rockerz 530 Foldable Bluetooth Headphone with Mic (Carbon Black)","1 Review"]
```

```python
scraper=AutoScraper()
result=scraper.build(yoshop_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(yoshop_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_adhv':'Title','rule_8mbr':'Price'})
scraper.keep_rules(['rule_adhv','rule_8mbr'])
scraper.save('yoshop-search')
```

```python
results=scraper.get_result_similar('https://yoshops.com/products?keywords=Boat',group_by_alias=True)
```
