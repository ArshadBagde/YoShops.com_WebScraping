```python
from autoscraper import AutoScraper
```

```python
yoshop_url="https://yoshops.com/t/fashion"
wanted_list=["₹399.00","QUECHUA Mini Laptop Bag 10 Liter- Red/Black","1 Review"]
```

```python
scraper=AutoScraper()
result=scraper.build(yoshop_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(yoshop_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_rfdl':'Title','rule_z6jl':'Price'})
scraper.keep_rules(['rule_rfdl','rule_z6jl'])
scraper.save('yoshop-search')
```

```python
results=scraper.get_result_similar('https://yoshops.com/t/fashion',group_by_alias=True)
```
