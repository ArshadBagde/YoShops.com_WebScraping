```python
from autoscraper import AutoScraper
```

```python
yoshop_url="https://yoshops.com/t/electronics"
wanted_list=["₹5,999.00","MitSonic 20 Inches (53 cm) FULL HD Gorilla Glass LED TV","1 Review"]
```

```python
scraper=AutoScraper()
result=scraper.build(yoshop_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(yoshop_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_1gjx':'Title','rule_dn5i':'Price'})
scraper.keep_rules(['rule_1gjx','rule_dn5i'])
scraper.save('yoshop-search')
```

```python
results=scraper.get_result_similar('https://yoshops.com/t/electronics',group_by_alias=True)
```
