```python
from autoscraper import AutoScraper
```

```python
yoshop_url="https://yoshops.com/t/food"
wanted_list=["₹160.00","1KG Chicken Biryani (Chennai)","1 Review"]
```

```python
scraper=AutoScraper()
result=scraper.build(yoshop_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(yoshop_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_8zlp':'Title','rule_emjn':'Price'})
scraper.keep_rules(['rule_8zlp','rule_emjn'])
scraper.save('yoshop-search')
```

```python
results=scraper.get_result_similar('https://yoshops.com/t/food',group_by_alias=True)
```
