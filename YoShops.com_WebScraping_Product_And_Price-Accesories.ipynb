```python
from autoscraper import AutoScraper
```

```python
yoshop_url="https://yoshops.com/t/accessories"
wanted_list=["₹1,999.00","Zebronics External TV Tuner Card TL1010","1 Review"]
```

```python
scraper=AutoScraper()
result=scraper.build(yoshop_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(yoshop_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_j2ph':'Title','rule_q45d':'Price'})
scraper.keep_rules(['rule_j2ph','rule_q45d'])
scraper.save('yoshop-search')
```

```python
results=scraper.get_result_similar('https://yoshops.com/t/accessories',group_by_alias=True)
```
