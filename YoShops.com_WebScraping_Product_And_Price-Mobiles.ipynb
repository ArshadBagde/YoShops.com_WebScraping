```python
from autoscraper import AutoScraper
```

```python
yoshop_url="https://yoshops.com/t/mobile-phones"
wanted_list=["₹1,499.00","Reliance Jio Phone","1 Review"]
```

```python
scraper=AutoScraper()
result=scraper.build(yoshop_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(yoshop_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_av49':'Title','rule_61f5':'Price'})
scraper.keep_rules(['rule_av49','rule_61f5'])
scraper.save('yoshop-search')
```

```python
results=scraper.get_result_similar('https://yoshops.com/t/mobile-phones',group_by_alias=True)
```
