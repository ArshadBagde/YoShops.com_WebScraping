```python
from autoscraper import AutoScraper
```

```python
yoshop_url="https://yoshops.com/t/toys"

wanted_list=["₹8,999.00","Sony PlayStation PS2 Gaming Console 150 GB Hard Disk With 50 Games Preloaded(Black)","1 Review"]
```

```python
scraper=AutoScraper()
result=scraper.build(yoshop_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(yoshop_url,grouped=True)
```

```python
scraper.set_rule_aliases({'rule_ixp1':'Title','rule_6jqx':'Price'})
scraper.keep_rules(['rule_ixp1','rule_6jqx'])
scraper.save('yoshop-search')
```

```python
results=scraper.get_result_similar('https://yoshops.com/t/toys',group_by_alias=True)
```
